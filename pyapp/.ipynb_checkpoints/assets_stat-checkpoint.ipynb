# 资产组成分析

```python
# Stat
assets_names = {CATEG_STOCK:[], CATEG_FUND: [], CATEG_CASH: []}
assets_netv = {CATEG_STOCK:[], CATEG_FUND: [], CATEG_CASH: []}
for k, sub_accnt in main_accnt.sub_accounts.items():
    if sub_accnt.net_amounts > 1000:
        assets_names[sub_accnt.categ].append(sub_accnt.code_name)
        assets_netv[sub_accnt.categ].append(sub_accnt.net_amounts)
sum_values = [sum(assets_netv[CATEG_STOCK]), sum(assets_netv[CATEG_FUND]), sum(assets_netv[CATEG_CASH])]
sum_labels = ['股票', '基金', '现金']

# Plot Pie
plt.figure(figsize=(36, 9),  dpi=300)
ax = plt.subplot(1, 3, 1)
wdges, labels, autopct = plt.pie(sum_values, labels=sum_labels, autopct=lambda x: f'{x/100 * sum(sum_values):.0f}\n{x:.2f}%', textprops={'size': 16})
plt.setp(labels, fontsize=18)
plt.title('资产组成', fontsize=24)
ax = plt.subplot(1, 3, 2)
wdges, labels, autopct = plt.pie(assets_netv[CATEG_FUND], labels=assets_names[CATEG_FUND], autopct=lambda x: f'{x / 100 * sum_values[1]:.0f}\n{x:.2f}%', textprops={'size': 12})
plt.setp(labels, fontsize=18)
plt.title('基金持仓成份', fontsize=24)
ax = plt.subplot(1, 3, 3)
wdges, labels, autopct = plt.pie(assets_netv[CATEG_STOCK], labels=assets_names[CATEG_STOCK], autopct=lambda x: f'{x / 100 * sum_values[0]:.0f}\n{x:.2f}%', textprops={'size': 12})
plt.setp(labels, fontsize=18)
plt.title('股票持仓成份', fontsize=24)
plt.tight_layout()
```

# 资产历史收益分析

```python
plt.figure(figsize=(9, 18),  dpi=100)
MAX_VALUE=150000
MAX_COLOR=1
# 盈利
ax = plt.subplot(1, 2, 1)
unit_topics = ['盈利']
accu_y_profit = np.array([0])
for k, sub_accnt in main_accnt.sub_accounts.items():
    y = np.array([0])
    if sub_accnt.net_profits > 500:
        y[0] = sub_accnt.net_profits
        p = plt.bar(unit_topics, y, bottom=accu_y_profit, width=0.4, label=sub_accnt.code_name)
        plt.bar_label(p, labels=[f'{sub_accnt.net_profits:.0f} - {sub_accnt.code_name} - {sub_accnt.code}'], label_type='center')
        accu_y_profit += y
plt.legend()
plt.ylim(0, MAX_VALUE)
# 亏损
ax = plt.subplot(1, 2, 2)
unit_topics = ['亏损']
accu_y_loss = np.array([0])
for k, sub_accnt in main_accnt.sub_accounts.items():
    y = np.array([0])
    if sub_accnt.net_profits < -500:
        y[0] = -sub_accnt.net_profits
        p = plt.bar(unit_topics, y, bottom=accu_y_loss, width=0.4, label=sub_accnt.code_name)
        plt.bar_label(p, labels=[f'{-sub_accnt.net_profits:.0f} - {sub_accnt.code_name} - {sub_accnt.code}'], label_type='center')
        accu_y_loss += y
plt.ylim(0, MAX_VALUE)
plt.legend()
print(accu_y_profit - accu_y_loss)
```

# 入库

```python
import pandas as pd
from sqlalchemy import create_engine

df = main_accnt.profit_df
# 假设您的 DataFrame 名为 df
# net_values
# date
# 2020-11-02 0.000000
# 2020-11-03 0.000000
# 2020-11-04 0.000000
# ...

# 创建一个数据库引擎
# 注意：您需要将 username、password、host、port 和 database_name 替换为您的数据库信息
engine = create_engine('mysql+pymysql://root:qepwd1231@127.0.0.1:3306/db_qe')

# 由于您的 DataFrame 只有净值一列，我们需要创建 init_profit 和 final_profit 列
# 假设 init_profit 和 final_profit 应该与 net_values 相同
df['init_profit'] = df['net_values'].shift(1).fillna(0)
df['final_profit'] = df['net_values']

# 现在 df 有三列：date, init_profit, final_profit
# 我们需要将 date 列作为索引
df.reset_index(inplace=True)

# 将 DataFrame 写入 MySQL，假设表名为 sum_profit
# if_exists 参数可以是 'fail', 'replace' 或 'append'
# 'fail': 如果表已存在，则引发值错误
# 'replace': 如果表已存在，将其替换
# 'append': 如果表已存在，将数据附加到表中
df.to_sql('sum_profit', con=engine, index=False, if_exists='append')

print("Data inserted successfully")
```
